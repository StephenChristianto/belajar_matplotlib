# Simple Pie Plot

disini kita membuat 2 data set yang diberi nama kategori dan data

lalu memanggil plt.pie dengan menyertakan data , labels nya kategori 

memanggil autopct untuk menampilkan persentase

dan memanggil startangle untuk mengatur rotasi 



```python
kategori = ['A','B','C','D','E']
data = [215,130,245,210,225]
```

```python

plt.pie(data,
       labels = kategori,
       autopct='%1.1f%%',
       startangle=90)

plt.title('Contoh Pie Plot')

plt.show()
```

# Pengaturan Legend dan Color# 

disini kita memakai data set yang tadi namun menambah kan 1 variabel warna 

```python
kategori = ['A','B','C','D','E']
data = [215,130,245,210,225]
warna = ['pink','cyan','#e03364','yellowgreen','skyblue']
```

```python
plt.pie(data,
       labels=kategori,
       colors=warna,
       autopct='%1.1f%%',
       startangle=90)

plt.legend()
plt.title('Contoh Pie Plot')

plt.show()
```

# Exploded Pie Plot

disini kita memakai data set yang tadi namun menambahkan 1 variabel ya itu explode_var

```python
kategori = ['A','B','C','D','E']
data = [215,130,245,210,225]
warna = ['pink','cyan','#e03364','yellowgreen','skyblue']
explode_var = [0., 0., 0.2, 0., 0.08]
```

```python
plt.pie(data,
       labels=kategori,
       colors=warna,
       explode=explode_var,
       autopct='%1.1f%%',
       startangle=90)

plt.title('Contoh Exploded Pie Plot')

plt.show()
```
