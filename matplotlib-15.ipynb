```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

# Sample Dataset

1.membuat variabel dengan menggenerate sebuah angka yang nilai terendah -3 dan max nya 3 sebanyak 100 data

```python
n = np.linspace(-3.0,3.0,100)

x,y = np.meshgrid(n,n)
z = np.sqrt(x**2+y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

# 3D Wireframe

1.memanggil plt.figure untuk membuat object figure

2.membuat object axes dengan menyertakan parameter projection bernilai 3d

3.memanggil fungsi ax.plot_wireframe

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_wireframe(x,y,z,color='orange')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Wireframe')

plt.show()
```

# 3D surface Plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x,y,z,color='orange')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Surface Plot')

plt.show()
```
