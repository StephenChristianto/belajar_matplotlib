# simple line plot

untuk generate data nya kita menggunakan fungsi np.arange()

fungsi np.arange diikuti dengan 3 parameter, yaitu:

Parameter pertama, yaitu 0.0 mengindikasikan starting point 

Parameter kedua, yaitu 2.0 mengindikasikan ending point 

Parameter ketiga, yaitu 0.01 untuk menetukan interval data.

```python
x = np.arange(0.0, 2.0, 0.01)
x
```

Data set kedua berisikan data dari nilai hasil operasi sin data set pertama.

  nilai x disini merupakan nilai hasil perkalian

```python
s = np.sin(2 * np.pi * x)
s
```

disini kita memanggil plt.subplots() yang akan menghasilkan dua buah nilai yang memerlukan dua variabel untuk menampung denganmenggunakan fig untuk figure dan ax untuk axes.

sumbu x akan merepresentasikan nilai yang ditampung oleh variabel x.

sumbu y akan merepresentasikan nilai yang ditampung oleh variabel s.

Untuk memberikan label kepada 3 nilai xlabel, ylabel, dan title secara sekaligus dapat menggunakan method ax.set() dengan diikutii oleh 3 parameternya.

Pemanggilan ax.grid() untuk menampilkan grid pada plottingnya.

```python
fig, ax = plt.subplots()

plt.plot(x,s)
ax.set(xlabel = 'nilaix',
       ylabel = 'nilai sin(x)',
       title = 'Visualisasi nilai sin')
ax.grid()

plt.show
```

# Multiple subplots

Multiple subplot merupakan dimana terdapat dua axes di suatu figure

pertama kita mempersiapkan data point :

1. terdapat 4 variabel yaitu x1,x2,y1,y2. yang berfungsi menampung 4 buah deret bilangan

2. nilai x1 dan x2 dihasilkan oleh pemanggilan fungsi np.linspace() yang disertai beberapa parameter :
        
        a.arameter pertama untuk mengatur nilai minimum (bersifat inclusive)
        b.Parameter kedua untuk mengatur nilai maximum (bersifat inclusive)
        c.Parameter ketiga untuk mengindikasikan jumlah data point yang akan dibuat

```python

x1 = np.linspace(0.0, 5.0, 100)
x2 = np.linspace(0.0, 2.0, 100)

y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
```

untuk menampilkan data yang kita buat denagn memanggil atau menulis x1

```python
x1
```

#  Multiple Subplots dengan OO Style

kita bisa menggunakan fungsi  plt.subplots() dengan disertai 2 parameter yaitu (2,1) yang berarti kita akan membuat 2 buah baris dan 1 kolom

lalu untuk menampung hasil plotting kita menyertai 2 variabel yaitu fig yang menampung figure dan (ax1, ax2) dan yang akan menampung 2 axes, ax1 dan ax2.

untuk menyertakan judul kita bisa memanggil fig.suptitle(Multiple Subplots).

untuk melakukan plotting pertama pada axes kita bisa menggunakan ax1.plot(x1, y1, 'ro-'). dengan parameter ro

untuk menentukan Nilai cos(x_1) pada axes pertama kita bisa menggunakan ax1.set_ylabel('Nilai cos(x_1)')

untuk melakukan plotting pada axes kedua kita bisa menggunakan ax2.plot(x2, y2, 'g.-') dengan parameter g.-

untuk menentukan Nilai cos(x_2) pada axes pertama kita bisa menggunakan ax2.set_ylabel('Nilai cos(x_2)')

lalu kita memangil ax2.set_xlabel('Nilai x') untuk  menampilkan nilai  'Nilai x' pada ax2.

```python

fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('Multiple Subplots')

ax1.plot(x1, y1, 'ro-')
ax1.set_ylabel('Nilai $cos(x_1)$')

ax2.plot(x2, y2, 'g.-')
ax2.set_ylabel('Nilai $cos(x_2)$')

ax2.set_xlabel('Nilai $x$')

plt.show()
```

# Multiple Subplots dengan pyplot Style

untuk mengendalikan axis 1 ada beberapa cara :
    1.memanggil plt.subplot() yang disertai parameter (2, 1, 1) yang artinya memiliki 2 kolom 1 baris dan merupakan axis ke-1
    
    2.melakukan plotting pada axis pertama dengan memanggil plt.plot(x1, y1, 'ro-') dengan parameter ro-
    
    3.untuk menyertakan judul figure dapat memanggil plt.title(Multiple Subplots).
    
    4.untuk menentukan label y yang menjadi 'Nilai cos(x_1)' pada axis pertama bisa memanggil plt.ylabel('Nilai cos(x_1)')
    
untuk mengendalikan axis 2 ada beberapa cara :

    1.memanggil plt.subplot()  yang disertai parameter (2, 1, 2) yang artinya memiliki 2 kolom 1 baris dan merupakan axis ke-2
    
    2.melakukan plotting pada axis kedua dengan memanggil plt.plot(x2, y2, 'g.-') dengan parameter g.-
    
    3.untuk menentukan y 'Nilai cos(x_2)'  pada axis pertama kita dapat menggunakan plt.ylabel('Nilai cos(x_2)')
    
    4. untuk menampilkan nilai  'Nilai x' pada ax2  kita memanggil plt.xlabel('Nilai x')

```python
plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'ro-')
plt.title('Multiple Subplots')
plt.ylabel('Nilai $cos(x_1)$')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'g.-')
plt.ylabel('Nilai $cos(x_2)$')
plt.xlabel('Nilai $x$')

plt.show()
```
