# 3D Line Plot

1.membuat 3 variabel , untuk variabel z untuk menampung np.random dari jangkauan 0 sampai 1 sebanyak 100 data

2.variabel x nya hasil z dikali np.sin 20 dikali z

3.variabel y nya hasil z dikali np.cos 20 dikali z

```python
z = np.linspace(0,1,100)
x = z * np.sin(20*z)
y = z * np.cos(20*z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot(x,y,z,color='red')

ax.set(title='Contoh 3D line plot',xlabel = 'sumbu X',ylabel = 'sumbu Y',zlabel = 'sumbu Z')
plt.show()
```

# Scatter plot

```python
z = np.linspace(0,1,100)
x = z * np.sin(20*z)
y = z * np.cos(20*z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter(x,y,z,color='red')

ax.set(title='Contoh 3D Scatter Plot',xlabel = 'sumbu X',ylabel = 'sumbu Y',zlabel = 'sumbu Z')
plt.show()
```
