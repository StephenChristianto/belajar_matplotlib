# Sample Data

1.Menyiapkan dua buah sampel yaitu mu dan sigma , mu berarti mean , sigma berarti standar deviasi

2.lalu menggunakan np.random.randn untuk agar menghasilkan jumlah angka yang telah di acak

3.agar mu nya 100 tidak 0 kita harus mengkalikannya dengan sigma np.random.randn(10000),lalu hasilnya dijumlahkan 

```python
mu, sigma = 100, 15 
x = mu + sigma * np.random.randn(10000)
x
```

# Histogram dengan pyplot style

1.menggunakan plt.hist dengan menyertakan 2 parameter , parameter wajib dan opsional

^ parameter wajib itu x dan bins = 50.

^parameter opsional itu facecolor = 'g' dan alpha = 0.75. , (facecolor = 'g' merepresentasikan green) ,(alpha = 0.75 untuk mengatur nilai transparasi menjadi 75%.)

2.memanggil plt.xlabel('Sumbu X'), plt.ylabel('Sumbu Y'), dan plt.title('Contoh Histogram')

3.memanggil plt.text() yang disertai oleh 3 parameter 

^koordinat x nya yaitu 45.

^koordinat y nya yaitu 500.

^yaitu '$\mu=100,\ \sigma=15$'

```python
plt.hist(x,
         bins=50,
         facecolor='g',
         alpha=0.75)

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogram')

plt.text(45, 500, r'$\mu=100,\ \sigma=15$')
plt.grid()

plt.show()
```

# Histogram Menggunakan OO Style

1.membuat 2 variabel yaitu fi dan ax , fig untuk menampung figure dan ax untuk menampung axes dari pemanggilan  plt.subplots().

2.memanggil  ax.hist() dengan menyertakan 2 parameter, yaitu parameter wajib dan  parameter opsional

^parameter wajib yaitu  x dan bins = 50.

^parameter opsional yaitu facecolor = 'g' dan alpha = 0.75.

3.memaggil ax.set_xlabel('Sumbu X'), ax.set_ylabel('Sumbu Y'), dan ax.set_title('Contoh Histogram') untuk menentukan judul pada histogram 

4.memanggil ax.text() yang disertai 3 parameter .

^pertama x nya yaitu 45.

^kedua y nya yaitu 500.

^ketiga yaitu '$\mu=100,\ \sigma=15$

5.Memanggil method ax.grid() untuk menampilkan grid pada plotting.

```python
fig, ax = plt.subplots()

ax.hist(x,
        bins=50,
        facecolor='r',
        alpha=0.75)

ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Contoh Histogram')

ax.text(45, 500, r'$\mu=100,\ \sigma=15$')
ax.grid()

plt.show()
```
