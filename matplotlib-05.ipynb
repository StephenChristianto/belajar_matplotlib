# Simpel bar plot 

1.membuat 2 buah data set yang diberi nama data dan kategori 

2.memanggil plt.bar dengan menyertakan kedua nama data set yang kita buat

3.kita membuat plt.xlabel dengan nama kategori 

4.membuat plt.ylabel dengan nama jumlah 

5.membuat plt.tittle yaitu contoh bar

```python

data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data)
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

# Pengaturan Grid dan Color

disini kita bisa mengubah atau men setting grid dan warna bar nya 

Mengubah warna pada saat penggunaan plt.bar() (nama data set) color = warna, alpha = transparansi

Mengubah grid dengan plt.grid()  linewidth = ketebalan grid, axis = arah grid sesuai sumbu

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python

plt.bar(kategori, data,
       color='red',
       alpha=0.25)

plt.grid(linestyle='--',
        linewidth=2,
        axis='y',
        alpha=0.50)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.xlabel('Contoh Bar Plot')

plt.show()
```

# Horizontal Bar Plot

disini kita menggunakan data yang sama 

lalu memanggil plt.barh dengan menyertakan nama dari data yang kita buat 

disini kita bisa menggunakan slicing juga 

kita membuat plt.xlabel dengan nama kategori 

membuat plt.ylabel dengan nama jumlah 

membuat plt.tittle yaitu contoh bar

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.barh(kategori, data)
#plt.barh(kategori[::-1],data[::-1])

plt.grid(linestyle='--',
        linewidth=1,
        axis='x',
        alpha=0.75)

plt.xlabel('Jumlah')
plt.ylabel('Kategori')
plt.title('Contoh Horisontal Bar Plot')

plt.show()
```

dengan fungsi slicing 

```python
plt.barh(kategori[::-1],data[::-1])

plt.grid(linestyle='--',
        linewidth=1,
        axis='x',
        alpha=0.75)

plt.xlabel('Jumlah')
plt.ylabel('Kategori')
plt.title('Contoh Horisontal Bar Plot')

plt.show()
```

# Stacked Bar Plot

disini kita membuat 3 data set 

stacked bar plot ini berfungsi untuk memplotting dua data set yang berbeda 

awalnya kita menggunakan fungsi plt.bar dengan menyertakan kategori dan data 1 dengan label data 1

lalu menggunakan fungsi plt.bar dengan menyertakan kategori dan data 2 yang diberi label data 2, dan memanggil bottom=data 1 

kita membuat plt.xlabel dengan nama kategori 

membuat plt.ylabel dengan nama jumlah 

membuat plt.tittle yaitu contoh bar


```python
data1 = [20, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A','B','C','D','E']
```

```python
plt.bar(kategori, data1, label='Data1')
plt.bar(kategori, data2, label='Data2', bottom=data1) #bottom : spesifikasi penempatan

plt.legend()

plt.grid(linestyle='--',
        linewidth=1,
        axis='y',
        alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```

# Grouped Bar Plot

disini kita membuat data yang sama , tetapi menambah variabel baru ya itu x untuk menampung array 

disini kita juga harus mendefinisikan lebar dari bar nya

```python

data1 = [20, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A','B','C','D','E']

x = np.arange(len(kategori))
x
```

```python
width = 0.35
plt.bar(x-width/2, data1, width, label='Data 1')
plt.bar(x+width/2, data2, width, label='Data 2')

plt.xticks(x, kategori)

plt.grid(linestyle='--',
        linewidth=1,
        axis='y',
        alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Grouped Bar Plot')

plt.legend()
```
