# membuat sampel data

disini kita membuat sampel data uas 

```python
data_uas=[
    ['Bejo', 70],
    ['Tejo', 83],
    ['Cecep', 62],
    ['wati', 74],
    ['Karti', 71]
]
```

```python
data_uas
```

# Tabel oo style

disini kita memanggil fungsi plt.subplot() yang menampung fig dan ax

lalu memanggil fungsi celltext dengan parameter data_uas dan menggunakan parameter loc dengan nilai center

lalu memngatur size tabel dengan memanggil table.scale

```python
fig, ax = plt.subplots()

table = plt.table(cellText=data_uas,
                 loc='center')

table.set_fontsize(14)
table.scale(0.5, 2) #ukuran kolom dan baris

ax.axis(False)

plt.show()
```

# pyplot style

disini kita tidak menggunakan fig dan ax namun selebihnya sama dengan script yang oos table

```python
table = plt.table(cellText=data_uas,
                 loc='center')

table.set_fontsize(14)
table.scale(1,4)

ax = plt.gca()
ax.axis(False)

plt.show()
```
