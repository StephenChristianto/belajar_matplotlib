# Kasus Twin Axes 1

memanggil fungsi np.linspace(1,10,25)  dengan menuliskan parameter yang akan menghasilkan data set dari 1 sampai 10

lalu melakukan twin plotting, plotting perama menggunakan plotting exponensial

dan plotting kedua akan menggunakan plotting logaritma

```python
x = np.linspace(1,10,25)
x
```

```python
fig,ax1 = plt.subplots()
ax1.plot(x,np.exp(x),'bs-',label='exp')
ax1.setx_label = ('Sumbu X')
ax1.sety_label = ('esp')

ax2 = ax1.twinx()
ax2.plot(x,np.log(x),'ro-',label='log')
ax1.sety_label = ('log')

fig.suptitle('Contoh Twin Axes')
fig.legend(loc = 'upper left')

plt.show()
```

# Kasus Twin Axes 2

1.menyiapkan sampel data set nya

2.lalu kita akan konversi data tersebut ke suhu celcius

```python
x = np.random.randint(16,40,size=30)
x
```

```python
def C2F(celcius=0):
    return (celcius*1.8)+32

def konversi_sumbu(ax1):
    y1,y2=ax1.get_ylim()
    ax2.set_ylim(C2F(y1),C2F(y2))
    ax2.figure.canvas.draw()
    
fig,ax1= plt.subplots()
ax2 = ax1.twinx()

ax1.callbacks.connect('y_lim changed',konversi_sumbu)
ax1.plot(x)

ax1.set_xlabel('Hari')
ax1.set_ylabel('Celcius')
ax2.set_ylabel('Fahrenheit')

fig.suptitle('Temperature Udara')
plt.show()
```
